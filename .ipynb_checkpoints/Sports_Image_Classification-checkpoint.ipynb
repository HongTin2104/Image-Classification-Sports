```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

```python
data_train_path = 'Sports_Data/train'
data_test_path = 'Sports_Data/test'
data_valid_path = 'Sports_Data/valid'
```

```python
img_width = 180
img_height = 180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle = True,
    image_size = (img_width, img_height),
    batch_size = 32,
    validation_split = False)
```

```python
data_categories = data_train.class_names
```

```python
data_valid = tf.keras.utils.image_dataset_from_directory(
    data_valid_path,
    shuffle = True,
    image_size = (img_width, img_height),
    batch_size = 32,
    validation_split = False
)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle = True,
    image_size = (img_width, img_height),
    batch_size = 32,
    validation_split = False
)
```

```python
plt.figure(figsize=(10,10))
for images, labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.title(data_categories[labels[i]])
        plt.axis('off')
```
